### 시계열 데이터 (Time Series)
- 시간의 흐름에 따라 기록된 정보들을 의미하며, 관측시간에 대한 관측자료의 관계로 표현한다.
- 시간 흐름에 따라 변화하는 방향성을 이해하고, 그 변화를 예측하는 것이 주 목적이다.
- 시계열 데이터의 유형에는 우연 변동 시계열, 계절 변동 시계열, 추세 변동 시계열, 계절-추세 변동 시계열 데이터가 있다.
- 자료의 인덱스가 시간인 데이터다.

#### 우연 변동 시계열
- 특정한 패턴이나 추세 없이 우연한 요소에 의해 변하는 데이터다.
- 패턴이나 규칙이 없고 우연히 발생하는 것처럼 보인다.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 우연 변동 시계열 데이터 생성
random1 = [1342, 1442, 1252, 1343, 
           1425, 1362, 1256, 1272, 
           1243, 1359, 1412, 1253, 
           1201, 1478, 1322, 1406, 
           1254, 1289, 1497, 1208]

# 시계열 데이터로 변환
# freq: 'D' 기본값 D는 1일을 의미한다.
# '5D'일 경우 5일 단위이고 
# '2W'일 경우 2주 단위이다.
# '3M'일 경우 3개월 단위이고,
# 분기별로 나타낼 때에는 'Q'를 사용한다(3월, 6월, 9월, 12월).
idx = pd.date_range(start='2019-01', periods=len(random1), freq='Q')
random1_ts = pd.DataFrame(random1, index=idx)

# 마지막 인덱스 5개 표시
display(random1_ts.tail())

# 시각화
plt.figure(figsize=(10, 6))
for col in random1_ts.columns:
    plt.plot(random1_ts.index, random1_ts[col], label=col)

plt.title('Random Variation Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

### 계절 변동 시계열
- 특정한 시간 주기(예시: 1달, 1년 등)에 따라 발생하는 패턴이 나타나는 데이터다.
- 특정 시간에 특정 사건이 반복된다.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 계절 변동 시계열 데이터 생성
season1 = [1142, 1242, 1452, 1543,
           1125, 1262, 1456, 1572,
           1143, 1269, 1462, 1553,
           1121, 1258, 1472, 1546,
           1154, 1249, 1477, 1548]

# 시계열 데이터로 변환
idx = pd.date_range(start='2019-01', periods=len(season1), freq='Q')
season1_ts = pd.DataFrame(season1, index=idx)

display(season1_ts.tail())

# 시각화
plt.figure(figsize=(10, 6))
for col in season1_ts.columns:
    plt.plot(season1_ts.index, season1_ts[col], label=col)

plt.title('Seasonal Variation Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

### 추세 변동 시계열
- 시간에 따라 일정한 방향으로 계속해서 증가하거나 감소하는 추세를 보이는 데이터다.
- 일정한 방향으로 계속해서 증가하거나 감소한다.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 추세 변동 시계열 데이터 생성
trend1 = [1142, 1242, 1252, 1343,
          1225, 1562, 1356, 1572,
          1343, 1459, 1412, 1453,
          1401, 1478, 1322, 1606,
          1554, 1589, 1597, 1408]

# 시계열 데이터로 변환
idx = pd.date_range(start='2019-01', periods=len(trend1), freq='Q')
trend1_ts = pd.DataFrame(trend1, index=idx)

display(trend1_ts.tail())

# 시각화
plt.figure(figsize=(10, 6))
for col in trend1_ts.columns:
    plt.plot(trend1_ts.index, trend1_ts[col], label=col)

plt.title('Trend Variation Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

### 계절-추세 변동 시계열
- 추세와 계절성 요인이 함께 나타나는 데이터로,  
  시간에 따른 일정한 추세와 특정 시간 주기에 따른 패턴이 동시에 나타나는 데이터다.
- 시간의 흐름에 따라 일정한 방향으로 움직이면서도 특정 시간에 특정한 일이 반복된다.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 계절-추세변동 시계열 데이터 생성
st1 = [1142, 1242, 1452, 1543,
       1225, 1362, 1556, 1672,
       1343, 1459, 1662, 1753,
       1421, 1558, 1772, 1846,
       1554, 1649, 1877, 1948]

# 시계열 데이터로 변환
idx = pd.date_range(start='2019-01', periods=len(st1), freq='Q')
st1_ts = pd.DataFrame(st1, index=idx)

display(st1_ts.tail())

# 시각화
plt.figure(figsize=(10, 6))
for col in st1_ts.columns:
    plt.plot(st1_ts.index, st1_ts[col], label=col)

plt.title('Seasonal-Trend Variation Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
