### 금융 시장의 수익률 (Financial Market Return)
- 일반적으로 금융 시장을 분석하거나 머신러닝 모델을 구축할 때  
  price(수익금)가 아닌 return(수익률)을 활용하는 경우가 많다.
- 수익률을 통해 복리 계산과 연율화(1년 간의 성장률) 등을 위해 무수히 많은 반복 계산을 하게 되므로  
  (1 + return)으로 계산할 경우, 복잡하고 번거로워진다.
- 기존 return에 log를 취해주면, 더 편한 계산이 가능해지지만 당연히 약간의 오차가 발생하게 된다.
- 하지만 약간의 오차를 감수하고 log를 사용함으로써 얻는 이득이 훨씬 많고,  
  국내 주식에서는 상한선과 하한선이 -30% ~ 30% 제약까지 있기 때문에 일반적인 주가 변동에 대해 오차가 극히 적다.
- 즉, log를 취하는 것은 정밀성보다 편의성을 높인 것으로 이해하면 된다.
- 수익률 단위가 분, 초, 밀리초 단위 이하로 내려가야 정밀성에 차이가 많이 발생하지만,  
  실제 금융 시장에서는 최소 하루 단위 이상으로 계산하기 때문에 오차가 거의 발생하지 않게 된다.
- 또한 return에 log를 취하면 우측으로 치우친 확률 분포를 중심으로  
  재조정해주는 효과까지 있기 때문에 안 쓸 이유가 없다.

```python
import yfinance as yf

# 이 중, VIX는 투자자들이 향후 30일 간 S&P 500 지수가 얼마나 변동할 것인지를 반영한 지수다
# 투자자들이 투자 결정을 내리기 전에 시장 리스크, 공포 및 스트레스를 측정하는 데 사용한다
columns = ['AAPL', 'MSFT', 'INTC', 'AMZN', 'GS', 
           'SPY', '^SPX', '^VIX', 'EURUSD=X', 
           '^XAU', 'GDX', 'GLD']

# 리스트 형태로 여러 개를 전달해서, 해당 주식들에 대한 데이터들을 전부 받아오는 것도 가능
f_df = yf.download(columns, start='2010-01-01')['Adj Close'].round(4)
f_df
```

```python
# shift: 행(인덱스 제외)을 n칸 아래로 내리고, 음수면 올린다
# 이번에는 이전 라그 대비 수익률을 구하기 위해 사용했다
display(f_df.shift(1).head(4))
display(f_df.head(4))
display(f_df.shift(-1).head(4))
```

```python
import numpy as np

# 수익률 - 다음 시간대의 수익으로 나눠준 다음, 로그를 취한다
rate_f_df = np.log(f_df / f_df.shift(1))
rate_f_df
```

```python
# 위 주식 종목들의 수익률을 하나의 플롯에 시각화
rate_f_df[['AAPL', 'MSFT', 'INTC', 'AMZN', 'SPY', 'GLD']].plot(figsize=(10, 5), lw=0.5)
```

```python
# 연율화
# 252는 연간 영업일 수(약 252일)
rate_f_df.mean() * 252
```

```python
# 결측치 제거
rate_f_df = rate_f_df.dropna()
rate_f_df
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 각 feature의 VIF Score(공분산성, 다른 feature와의 상관관계)를 구해주는 함수
# 이번에는 각 종목 별로 다른 종목과의 상관관계가 어느 정도인지를 구하기 위해 사용했다
def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
# 각 주식 종목의 VIF Score 출력
get_vif(rate_f_df)
```

```python
# 각 주식 종목의 데이터 분포를 히스토그램으로 출력
rate_f_df.hist(figsize=(10, 10), bins=50)
plt.show()
```

```python
# 일간 수익률
# 수익률 산출할 때 로그를 취했으니, 지수를 취해 값을 원복시킨다
# cumsum: 각 원소의 누적합 (cumulative sum)
# 누적되는 일련의 과정을 list에 담아준다
rate_f_df.cumsum().apply(np.exp).plot(figsize=(12, 6))
plt.show()
```

```python
# 월간 수익률
# 매 달(1m)의 마지막(last()) 별 누적합을 시각화
rate_f_df.cumsum().apply(np.exp).resample('1m').last().plot(figsize=(12, 6))
plt.show()
```

```python
# 아마존(Amazon)의 날짜 별 주가만 가져오기
amzn_df = f_df[['AMZN']].dropna()
amzn_df
```

```python
window_size = 20

# 각 라그 별로 다음 20 라그(영업일)까지의 최소값, 평균, 편차, 중간값, 최대값을 새로운 컬럼으로 생성
amzn_df['min'] = amzn_df['AMZN'].rolling(window=window_size).min()
amzn_df['mean'] = amzn_df['AMZN'].rolling(window=window_size).mean()
amzn_df['std'] = amzn_df['AMZN'].rolling(window=window_size).std()
amzn_df['median'] = amzn_df['AMZN'].rolling(window=window_size).median()
amzn_df['max'] = amzn_df['AMZN'].rolling(window=window_size).max()

amzn_df.dropna()
```

```python
import matplotlib.pyplot as plt

# 하한가, 평균, 상한가 시각화
# iloc[-252:]로 설정한 이유는, 최근 1년간 영업일의 주가 변동을 시각화하기 위함
ax = amzn_df[['min', 'mean', 'max']].iloc[-252:].plot(figsize=(12, 6), style=['g--', 'r--', 'g--'], lw=0.8)
amzn_df['AMZN'].iloc[-252:].plot(ax=ax)
plt.title('AMZN 20-Day Moving Average Price Movement')
plt.show()
```

```python
# SMA(Simple Moving Average): 일정 기간 동안의 가격 평균을 나타내는 보조 지표
# 월간 영업일을 20일, 연간 영업일을 252일로 가정

# 최근 1개월간, 최근 1년간 영업일의 주가 평균을 새로운 컬럼으로 추가 
amzn_df['SMA1'] = amzn_df['AMZN'].rolling(window=21).mean() # short-term
amzn_df['SMA2'] = amzn_df['AMZN'].rolling(window=252).mean() # long-term
amzn_df[['AMZN', 'SMA1', 'SMA2']].tail()
```

```python
# 각 라그의 주가, 월간, 연간 평균 시각화
amzn_df[['AMZN', 'SMA1', 'SMA2']].plot(figsize=(12, 6))
plt.title("AMZN Simple Moving Average - Long & Short-term")
plt.show()
```

```python
# 아마존 주가 기술 분석
# 골든 크로스, 데드 크로스
# 월 평균 > 연 평균이 되는 지점이 골든 크로스
# 월 평균 < 연 평균이 되는 지점이 데드 크로스다
amzn_df.dropna(inplace=True)

# 골든 크로스 / 데드 크로스 기준에 따라 다른 값을 가진 새로운 컬럼 생성 (매도 및 매수 시기)
# 언제 사고, 언제 팔아야 되는지를 알 수 있는 지표다
amzn_df['positions'] = np.where(amzn_df['SMA1'] > amzn_df['SMA2'], 1, -1)  # 1: buy , -1: sell /

# 각 라그 별 주가, 월 평균, 연 평균, 매도 / 매수 시기 시각화
ax = amzn_df[['AMZN', 'SMA1', 'SMA2', 'positions']].plot(figsize=(15, 6), secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((-0.05, 1))

plt.title("AMZN Trading Window based on Technical Analysis")
plt.show()
```

### 24/06/12 - auto_arima

```python
import pandas as pd

# financial_market.csv 파일 읽어오기
# index_col: 해당 컬럼을 인덱스로 지정
# parse_dates: 인덱스 컬럼을 datetime 형식으로 변경
f_df = pd.read_csv('./datasets/financial_market.csv', index_col='Date', parse_dates=True)
f_df
```

```python
# pip install pmdarima
```

```python
import pmdarima as pm
```

```python
# 결측치 제거
pre_f_df = f_df.dropna()
pre_f_df
```

```python
# 최근 1년 간 영업일의 주가만 가져오기
amzn_df = pre_f_df[['AMZN.O']].iloc[-252:]
amzn_df
```

```python
# train과 test를 8:2 비율로 분할 (앞쪽 80%:뒤쪽 20%)
y_train = amzn_df['AMZN.O'][:int(0.8 * len(amzn_df))]
y_test = amzn_df['AMZN.O'][int(0.8 * len(amzn_df)):]
```

```python
# train, test 시각화
y_train.plot()
y_test.plot()
```

```python
from pmdarima.arima import ndiffs

# KPSS(Kwaiatkowski-Phillips-Schmidt-Shin)
# 차분을 진행하는 것이 필요할지 결정하기 위해 사용하는 한 가지 검정 방법
# 영가설(귀무가설)을 "데이터에 정상성이 나타난다"로 설정한 뒤
# 영가설이 거짓이라는 증거를 찾는 알고리즘이다
kpss_diffs = ndiffs(y_train, alpha=0.05, test='kpss', max_d=6)
adf_diffs = ndiffs(y_train, alpha=0.05, test='adf', max_d=6)
pp_diffs = ndiffs(y_train, alpha=0.05, test='pp', max_d=6)

# 위의 3가지 방식으로 최적 차분 수를 구한 뒤, 그 중 최대 값을 구한다
# 여기서 나온 차수를 아래의 auto_arima의 차분 수(d) 파라미터에 할당하면 된다
n_diffs = max(kpss_diffs, adf_diffs, pp_diffs)

print(f'd = {n_diffs}')
```

```python
import pmdarima as pm

# auto_arima
# GridSearch같이 최적의 파라미터를 구해준다
# 단, 모든 경우의 수를 계산하는 것이 아니라 파라미터 별로 순차적으로 구해나간다(stepwise=true)
# 그 과정에서 모델의 성능이 갱신될 때만 해당 모델의 파라미터를 출력하며
# 최종적으로 가장 좋은 성능을 보였을 때의 파라미터로 ARIMA 모델을 자동 세팅하며, 그 때의 파라미터 값도 출력한다 

# d: 차분 수
# start_p, max_p: AR 차수. 아래 코드는 0 ~ 10차 사이의 값을 계산하며 최적의 성능을 찾는다
# start_q, max_q: MA 차수. 마찬가지로 0 ~ 10차 사이의 값을 계산하며 최적의 성능을 찾는다
# m: 주기(개월)
# seasonal: 계절성(default=False). True로 설정할 시, m개월마다의 패턴을 찾는다 - 제대로 알아볼 것!
# stepwise: True로 설정할 시, 모든 경우의 수를 전부 고려하는 것이 아닌 파라미터 순서(d → p → q)로 최적의 파라미터 값을 구해나간다
# trace: True로 설정할 시, 성능이 갱신되었을 때의 파라미터를 출력한다
model = pm.auto_arima(y=y_train,
                      d=1,
                      start_p=0, max_p=10,
                      start_q=0, max_q=10,
                      m=1, seasonal=False,
                      stepwise=True,
                      trace=True)
```

```python
model.fit(y_train)
```

```python
# Prob(Q), 융-박스 검정 통계량 
# 영가설: 잔차가 백색잡음 시계열을 따른다
# 0.05 이상: 서로 독립적이고 동일한 분포를 따른다

# Prob(H), 이분산성 검정 통계량
# 영가설: 잔차가 이분산성을 띠지 않는다
# 0.05 이상: 잔차의 분산이 일정하다

# Prob(JB), 자크-베라 검정 통계량
# 영가설: 잔차가 정규성을 따른다
# 0.05 이상: 일정한 평균과 분산을 따른다

# Skew (왜도, 분포가 쏠려있는 정도)
# 0에 가까울수록 좋다

# Kurtosis (첨도, 분포가 뾰족한 정도)
# 3에 가까울수록 좋다

print(model.summary())
```

```python
import matplotlib.pyplot as plt

# 모델의 잔차, 분포(히스토그램, 산점도), ACF 시각화
# 우상단의 히스토그램에서 음수쪽 분포가 쏠려 있는 형태를 띤 것이
# 좌하단의 산점도에서는 y=x 그래프에서 다소 벗어난 분포를 보이는 형태로 나타났다
model.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
# 시계열 데이터를 predict할 때는 데이터가 아닌 주기(테스트 데이터의 길이)를 넣어준다
# 단, 아래 코드대로 predit 한 것은 정확한 예측이 아닌데
# 그 이유는 한 번 예측할 때마다 업데이트가 이루어지지 않았기 때문이다
# 아래의 예측 결과를 보면, 예측값에 대한 업데이트가 이루어지지 않아 등차수열 형식을 보이는 것을 알 수 있다
prediction = model.predict(len(y_test))
prediction
```

```python
# 2라그에 대한 예측과 신뢰구간(confidence interval)을 predict하고 출력
prediction, conf_int = model.predict(n_periods=2, return_conf_int=True)
print(conf_int)
print(prediction)
```

```python
prediction.to_list()[0]
```

```python
# 각 라그 별 예측 결과를 반환하는 함수
def predict_one_step():
    prediction = model.predict(n_periods=1)
    return prediction.tolist()[0]
```

```python
p_list = []

# test 데이터의 각 라그 별 예측 결과를 리스트에 담고,
# 각 예측이 끝날 때마다 모델 업데이트
for data in y_test:
    p = predict_one_step()
    p_list.append(p)

    model.update(data)
```

```python
# 각 라그 별 실제 test 데이터와 모델의 예측값을 데이터프레임으로 생성
y_predict_df = pd.DataFrame({"test": y_test, "pred": p_list})
y_predict_df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

# train, test 데이터와 모델의 예측값을 그래프로 시각화
plt.plot(y_train.iloc[-50:], label='Train')
plt.plot(y_test.iloc[-50:], label='Test')
plt.plot(y_predict_df.pred, label='Prediction')
plt.legend()
plt.show()
```

```python
import numpy as np

# 모델의 예측과 실제 데이터 간 오차(%)를 구해주는 함수
def MAPE(y_test, y_pred):
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100

# 모델의 예측과 실제 데이터 사이의 오차(%) 출력
print(f'MAPE (%): {MAPE(y_test, p_list):.4f}')
```
