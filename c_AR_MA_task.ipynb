### ARIMA task

```python
import pandas as pd
import json

with open('./datasets/bitcoin_2010_2024.json') as f:
    json_data = json.load(f)

bitcoin_df = pd.DataFrame(json_data['market-price'])
```

```python
bitcoin_df
```

```python
from datetime import datetime

def changeDate(milis):
    timestamp = milis / 1000
    converted_time = datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')

    return converted_time
```

```python
bitcoin_df.x = bitcoin_df.x.apply(changeDate)
bitcoin_df
```

```python
bitcoin_df.set_index('x', inplace=True)
bitcoin_df
```

```python
bitcoin_df.info()
```

```python
bitcoin_df.index = pd.to_datetime(bitcoin_df.index)
bitcoin_df.info()
```

#### 데이터 시각화

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20, 5))

plt.subplot(1, 3, 1)
plt.plot(bitcoin_df, c='orange')
plt.title('Bitcoin Price(2010~2024)')

plt.subplot(1, 3, 2)
plt.plot(bitcoin_df.diff())
plt.title('Bitcoin Differencing of Price(2010~2024)')

plt.subplot(1, 3, 3)
plt.plot(bitcoin_df.pct_change())
plt.title('Bitcoin Percent of Price(2010~2024)')

plt.show()
```

#### ACF, PACF 확인

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# ACF
plot_acf(bitcoin_df, lags=20, ax=ax[0])
plot_acf(bitcoin_df.diff().dropna(), lags=20, ax=ax[1])
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# PACF
# 처음으로 0에 수렴한 이후, 다시 영역 밖으로 나간 절단점의 x값 = 대개 해당 차수로 차분하면 예측 성능이 좋아진다
plot_pacf(bitcoin_df, lags=20, ax=ax[0])
plot_pacf(bitcoin_df.diff().dropna(), lags=20, ax=ax[1])
plt.show()
```

```python
import numpy as np
from statsmodels.tsa.stattools import acf

bitcoin_df.dropna(inplace=True)
bitcoin_diff_df = bitcoin_df.diff().dropna()

bitcoin_acf = acf(bitcoin_df, nlags=20)
bitcoin_diff_acf = acf(bitcoin_diff_df, nlags=20)

plt.figure(figsize=(12, 5))

# 차분 이전의 자기 상관 함수(ACF, Autocorrelation Function)
plt.subplot(121)
plt.plot(bitcoin_acf, c='orange')
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96 / np.sqrt(len(bitcoin_df)), linestyle='--', color='gray')
plt.axhline(y=1.96 / np.sqrt(len(bitcoin_df)), linestyle='--', color='gray')
plt.title('Non-Stationarity Autocorrelation Function')

# 차분 이후의 자기 상관 함수
# 차분 이후에는 평균과 분포가 안정된다
# 만약 1차 차분 이후에도 값이 점선 범위에서 많이 벗어난다면 한 번 더 차분해준다
# 아래 그래프에서 값이 처음으로 파란 영역 안에 들어가는 2로 차분하는 것이 가장 좋다는 것을 알 수 있다
plt.subplot(122)
plt.plot(bitcoin_diff_acf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96 / np.sqrt(len(bitcoin_diff_df)), linestyle='--', color='gray')
plt.axhline(y=1.96 / np.sqrt(len(bitcoin_diff_df)), linestyle='--', color='gray')
plt.title('Stationarity Autocorrelation Function')

plt.show()
```

#### 중간 분석
- 비트코인 가격은 증가 추세
- ACF 그래프 상의 절단점은 x(라그)가 2일 때이므로, AR모델의 차수는 1로 지정
- 원본 비트코인 가격 그래프의 패턴이 뾰족하기 때문에, MR은 20으로 지정

#### ARIMA 모델로 예측

```python
from statsmodels.tsa.arima.model import ARIMA

# ARIMA(데이터, order=(AR, 차분, MA))
model = ARIMA(bitcoin_df, order=(6, 1, 21))

# ARIMA 모델은 fit 한 결과를 반드시 변수에 저장할 것!
model_fit = model.fit()

start_index = pd.to_datetime('2018-01-04')
end_index = pd.to_datetime('2024-01-01')

# fit한 모델 객체로 바로 predict
forecast = model_fit.predict(start=start_index, end=end_index)

# 원본 그래프와 모델의 예측값 시각화
plt.figure(figsize=(15, 8))
plt.plot(bitcoin_df, label='original')
plt.plot(forecast, label='predicted', c='orange')
plt.title("Time Series - Bitcoin Price Forecast")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_squared_log_error
import numpy as np

mse = mean_squared_error(bitcoin_df['2018-01-04':'2024-01-01'], forecast)
print('Mean Squared Error:', mse)

msle = mean_squared_log_error(bitcoin_df['2018-01-04':'2024-01-01'], forecast)
print('Mean Squared Log Error:', msle)
```
